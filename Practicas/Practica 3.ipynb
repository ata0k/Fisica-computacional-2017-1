## Cuarto paso, configuración inicial.

Para poder iniciar tus simulaciones, necesitas tener una configuración inicial, la cual en principio debe ser aleatoria. Es decir, que sus velocidades sean aleatorias y sus posiciones también. Para las velocidades no hay mucho problema, simplemente se requieren $n$ números aleatorios $\theta_i$ entre 0 y $2 \pi$ y aplicar $v_i = [cos(\theta_i),sin(\theta_i)]$. Para las posiciones iniciales es más complicado, pues no queremos que las partículas se encimen inicialmente. Para esto, se tendrán que ir poniendo una por una. 

[8] Haz una función que, dado un arreglo de $n$ posiciones iniciales con sus respectivos radios, agregue una nueva posición aleatoria que no se encime con las anteriores. Para esto tendrás que usar un "while".

# Parte 2: Método de Monte Carlo

Lo más adecuado es en general hacer dinámica molecular, lo cual es de alguna forma "determinístico" y se pueden incluir, sin mayor dificultad otros potenciales o características (por ejemplo, discos rotando); sin embargo, es mucho más complejo y no siempre se puede hacer debido al tiempo de cómputo. 

En esta parte desarrollaremos el método de Monte Carlo para estudiar discos (o esferas) duros(as). La idea es realmente muy simple. Se desplazan las partículas en dirección aleatoria una cantidad $\delta$ pequeña. Si al desplazar las partículas estas no se enciman, entonces se dejan estas nuevas posiciones. En cambio, si lo hacen, se regresa a la posición inicial con cierta probabilidad y se vuelve a aplicar el método.

Igual que en el caso de Ising, se puede aplicar esto a diferentes potenciales (no sólo disco duro), y el orden en el que se elige la partícula que se moverá es aleatorio. Entonces, el siguiente paso es poder reconocer cuando 2 discos se enciman.

# Parte 3: Análisis. 

Ahora estamos interesados en hacer algo de análisis. Lo que nos interesa es observar cuándo hay una transición de fase. Para esto podríamos medir cambios en la presión dados cambios en la densidad, pero eso es demasiado difícil por ahora. Lo que tienen que hacer es observar la configuración final de las partículas después de suficientes colisiones. 

En una trancisión de fase, suele haber un cambio en el orden del sistema, pasando de "desordenado" a "ordenado". Típicamente lo que se obtiene es que aparece una simetría rotacional y una traslacional. En el caso de los discos duros, esto se traduce en tener las particulas acomodadas de forma aleatoria o bien, en un arreglo cuadrático o hexagonal.

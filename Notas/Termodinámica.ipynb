# Escalas termodinámicas

En física utilizamos una escala muy particular llamada la escala de Kelvin. 

En el siglo XIX, el físico inglés, Lord Kelvin, asumiendo la hipótesis molecular de los gases, calculó un límite de la temperatura a la cual un gas ideal tendería a tener un volumen 0. Su cálculo lo hizo en base a experimentos hechos sobre gases nobles (como el helio, o el neon). 

Lo que descubrió fue que existe una temperatura mínima llamada el 0 absoluto. La escala de Kelvin tomó como base la de Celcius, que divide en 100 grados la diferencia entre el punto triple del agua (punto al cual el agua puede encontrarse en 3 estados) y el punto de ebullición del agua a esa misma presión. La única diferencia es que en el caso de la escala Celcius, el punto triple se encuentra en $0^o C$, mientras que la de Kelvin, los 0 grados corresponden al punto donde las partículas perden toda su energía. La diferencia es $0K = -273.16^o C$. 

Además de estas dos escalas, existen otras, como la escala Farenheit, basadas en datos extraños. En el caso de Farenheit, la medida fue la temperatura de su esposa, que consideró los $96^o F$ y la temperatura más baja de su pueblo natal, la cual consideró $0^o F$. Tú podrías hacer tu propia escala de la temperatura. Así lo hicieron. En la antigüedad hubo varios físicos que hicieron sus propias escalas. 

Para no batallar con las escalas (en especial con la Farenheit), haremos un función para poder hacer todos los cálculos de forma rápida. 

```julia
function Temperatura(T, Sistema1, Sistema2) 
    if Sistema1 == "F" && Sistema2 == "C"
        return (T-32)*5/9
    elseif Sistema1== "C" && Sistema2 == "F"
        return 9/5 * T + 32
    elseif Sistema1== "C" && Sistema2 == "K"
        return T + 273.16
    elseif Sistema1== "K" && Sistema2 == "C"
        return T - 273.26
    elseif Sistema1== "K" && Sistema2 == "F"
        return (T - 273.26)*9/5 +32
    elseif Sistema1== "F" && Sistema2 == "K"
        return (T-32)*5/9+ 273.16
    elseif Sistema1==Sistema2
        return T
    else
        return "No entendí de qué sistema a qué sistema quieres convertir"
    end
end
```

# Gas ideal

Un gas, donde las partículas no interactuan entre ellas, se conoce como un gas ideal. Estudiar un gas como este resulta de gran importancia, porque, aunque es tan solo una aproximación a la realidad, es bastante acertada en la gran mayoría de los casos. 

En un gas típico, la densidad es relativamente baja, lo que quiere decir que, cada partícula ocupa un espacio muy grande, por lo que la interacción entre moléculas es prácticamente nula. Pensemos por ejemplo en el aire. 

El aire tiene una densidad media de $\rho_{air} = 1.2922 kg/m^3$. 

Problema 1: ¿Cuántas moléculas hay en un metro cúbico de Aire? Considera que el aire está constituido básicamente por 75% de nitrógeno y 25% de Oxígeno ($O_2$). Una mol de Nitrógeno tiene una masa de 14g, mientras que uno de $O_2$ tiene una masa de 32g. 

Problema 2: ¿Cuánto espacio hay por cada molécula? 

Problema 3: si en promedio cada molécula mide aproximadamente $0.7x10^{-9}m$ de radio, ¿Cuál es la probabilidad de que 2 moléculas interaccionen (se encimen)?

```julia
# hay 1292.2g de masa en un metro cúbico. Llamemos M a la masa molar de la combinación de los gases. 
M = .75*14+0.25*32
# Entonces, habrá 
n = 1292.2/M  #moles de aire. 
# Que multiplicado por el número de avogadro, obtenemos que hay 
N = n*6.022e23 # moléculas de aire en un metro cúbico. 
```

$4.206x 10^{25}$ moléculas parece muchísimo, pero ahora respondamos la pregunta 2. ¿Cuánto espacio tiene cada molécula?

Para esto dividimos 1, entre $N$ y con eso sabremos el espacio en $m^3$. Luego, valdría la pena pasarlo a nanómetros cúbicos, que es la unidad de un radio molecular promedio. Para esto, dividimos ese mismo resultado por (10^{-27}), es decir:

```julia
espacio = 1/N/(1e-27)
```

Para contestar la última pregunta, primero tenemos que calcular cual es el tamaño de una molécula promedio de aire. Para esto usamos la fórmula del volumen de una esfera, es decir: $\frac{4}{3} \pi r^3$. En nanometros esto es:

```julia
v = 4/3 * π * (.7*1e-9)^3/1e-27
# Ahora dividimos v espacio, para saber la densidad de particulas 
densidad = v/espacio
```

esta es la probabilidad de que 2 particulas colisionen. Como se ve, es una probabilidad bastante baja, por lo que podemos pensar que el aire se comporta casi como un gas ideal. Si además usáramos gases nobles, su radio de interacción se reduce a casi un décimo, por lo que la probabilidad de interacción se reduce unas mil veces. En esos casos la aproximación de gas ideal se cumple prácticamente en todas las condicciones. 

# ¿Y cómo se estudia un gas ideal?

En general cualquier sistema termodinámico se estudia mediante su ecuación de estado. Para obtener la ecuación de estado de un gas ideal utilizaron algo que se conoce como "física estadística" (que es a lo que yo me dedico). 

En la física estadística, se busca calcular todos los posibles estados en los que puede estar un sistema dada una energía. Luego, con eso se puede calcular la entropía del sistema, con lo que a su vez se calcula la temperatura, el volumen y la presión. En el caso del gas ideal lo que se obtiene es que el gas se comporta de la siguiente manera: 

$ P \cdot V = NKT$ 

Aquí $P $ es la presión, $V$ el volumen del gas, $N$ el número de partículas, $K$ es la constante de los gases y $T$ la temperatura. 

Usando esta ecuación (de forma aproximada y en grados celcius), o sea: $ P \cdot V = NKT- 273.16NK$, es que Kelvin encontró el cero absoluto. Esto, por supuesto, haciendo medidas experimentales con gases nobles.  


# Transferencia de Calor

El calor se puede trasmitir de 3 formas:

1. Por radiación. De la misma forma en la que la luz viaja, también el calor puede viajar, es decir, mediante fotones. En general estamos acostumbrados a que el calor se da por fotones infrarojos, pero en realidad cualquier onda electromagnética puede trasmitir energía y por lo tanto calor. 
- Por convección. 
- Por contacto. 

El primero de los puntos lo olvidaremos por el momento. Nos concentraremos en los últimos 2 puntos. 

## Convección:

Usando la ecuación del gas ideal podemos ver que, cuando un gas está a mayor temperatura, pero su presión se mantiene igual, necesariamente el volumen del gas cambia; sin embargo, su masa no cambia. Esto significa que el gas tendrá menor densidad al calentarse. 

Ahora bien, por el principio de Arquímides, 

$F_{flotacion} = \rho_{g_f} g V_c = m g \frac{V_c}{V_f}$,

lo que implica que el aire que se calienta sube con una aceleración $g-g \frac{V_c}{V_f} =g(1-\frac{V_c}{V_f})$

Ahora bien, si en la parte superior hay un sistema de refrigerado (por ejemplo, debido a la expanción del aire mismo), el aire empezará a circular, calentandose abajo y enfriandose arriba. 

## Conducción

Al tocar un objeto a diferente temperatura, el calor fluirá del objeto más caliente, al más frío. Esto es lo que se conoce como segunda ley de la termodinámica. 

Notese que escribí "fluirá calor". ¿Qué es el calor? El calor, de alguna forma, es una froma de energía. ¿Cómo podemos calcular el calor de un objeto? Para calcular el calor de un objeto, usamos 3 valores: 

1. La temperatura del objeto, 
- La masa del objeto
- El calor específico del objeto. 

El calor es entonces $Q = L_e \cdot m \cdot T$. 

Como mencioné antes, la segunda ley de la termodinámica asegura que no se pasará calor de cuerpos fríos a calientes, así que si ponemos 2 cuerpos en contacto, igualarán sus temperaturas, cuando ambos tengan cantidades de calor equivalente por cantidad de masa. 

Por ejemplo, el calor específico del agua es 4.18, el del hierro es 0.450. Supongamos que tenemos una sartén de hierro a $80^o C$ y la queremos enfriar con agua a $20^o C$. Si la sartén tiene una masa de 200g, ¿cuánta agua necesitaremos para que la sartén llegue a $22^o C$?

```julia
ΔQ1 = (80-22)*200*0.45
ΔQ2 = ΔQ1
# Entonces 2*m*4.18=ΔQ2, por lo que m= ΔQ2/(2*4.18)
m = ΔQ2/(2*4.18)
```

# Movimiento Browniano

Hace muchos años (en el siglo XIX) un biólogo inglés de apellido Brown, notó que cuando ponía un grano de polen en el agua, este seguía un movimiento en zigzag. El polen parecía tener decición propia, moviéndose asarozamente tal como lo hace una mosca en el aire. Su hipótesis fue justamente esa. 

Muchos años más tarde, Albert Einstein trató de explicar ese movimiento mediante la hipótesis molecular de Avogadro. Esta hipótesis sostiene que todos los componentes están conformados por miles de millones de moléculas, concretamente, $6.022 x 10^{23}$ moléculas por mol. Entonces, la idea de Einstein es que cada una de esas moléculas golpea de forma azarosa al grano de polen. El grano de polen es entonces golpeado cientos de miles de veces en diferentes direcciones por segundo en todas direcciones, por lo que termina teniendo un movimiento azaroso que se bautizó después como el caminante aleatoreo o bien, caminante borracho. 

En lo que sigue vamos a hacer una animación de como se vería este movimiento si la hipótesis de Einstein fuera verdadera. 

Para esto nos basaremos en un resultado del famoso físico Maxwell, en el cual calculó la velocidad de las moléculas en un fluido a una temperatura dada. 

$f(v) = \sqrt{\left(\frac{m}{2 \pi kT}\right)^3}\, 4\pi v^2 e^{- \frac{mv^2}{2kT}}$

Esta es la distribución de velocidades de un gas de partículas de masa m a una temperatura dada. Veamos, primero que nada, cómo se ve esta distribución para, por ejemplo, el agua a $T = 20^o C = 283.16K$

```julia
using Plots
```

```julia
k = 1.38064852e-23
av =  6.022e23
m = .018 / av
mk = m/k
T = 293.16
f(v, m, T) = √((mk/(2π*T))^3)*4*π* v.^2 .* exp(-mk*v.^2 ./(2*T))

using PyPlot
```

```julia
plot(0.1:0.1:3000,f(0.1:0.1:3000, m, T)) 
```

Vemos que la velocidad de las moléculas tiene un máximo al rededor de los $600m/s$ (o $2160km/h$). Sí, esa es la velocidad promedio a la que viajan las moléculas de agua. Rapidísimo, ¿cierto? Ok, pero apenas y tienen una masa $m = 2.989040185984722x10^{-26} kg$, por lo que su energía cinética es apenas de $.53802723347725x10^{-20} J$.

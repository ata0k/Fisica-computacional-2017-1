[16] ¿Qué tan rápida es la convergencia al valor correcto? Es decir, ¿cómo depende la distancia del iterado xn desde el límite, como función del número de paso n? Dibuja unos iterados.


```julia
aproximacion_raiz(x,b) = ( b + x/b )/2   #problema 13. 
```

```julia
function raiz(x,b,δ=1e-15)  #problema 14
    while abs(b^2 - x) > δ
        b = aproximacion_raiz(x,b)
    end
    return b
end
```

```julia
raiz(x) = raiz(x,1) #problema 15

raiz(5)-sqrt(5)
```

```julia
X = []    #Problema 16
Y = []
b = 1
for i = 1:20
    b = aproximacion_raiz(160000,b)
    push!(X,i)
    push!(Y,abs(b-400))
end
yscale("log")
xlabel("pasos")
ylabel(L"\Delta_b] \sqrt{160000}")
scatter(X,Y)
plot(1:20, 0.2e6*exp(-0.8*(1:20)))
```

Como se ve, inicialmente cae de forma exponencial, aunque después de algunos pasos la convercencia es incluso mayor.
